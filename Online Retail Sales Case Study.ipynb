# 1. Data Exploration and Cleaning

```python
# Load the dataset into a Pandas DataFrame and display the first 5 rows.

data = pd.read_csv("online_retail_sales_Large.csv")
data.head()
```

```python
# Check the dataset shape, column names, data types, and summary statistics.

print("Shape of Dataset \n", data.shape)
print("Columns Name \n", data.columns.to_list())
print("Data Type of Dataset \n", data.dtypes)
print("Summary Statistics \n", data.describe(include="all"))
```

```python
# Identify missing values in each column.

print(data.isnull().sum())
```

```python
# Handle missing values appropriately:
# Numerical columns: fill with mean or median.
# Categorical columns: fill with mode or drop.
```

```python
# Convert Order_Date into datetime format.

data["Order_Date"] = pd.to_datetime(data["Order_Date"])
```

```python
# Extract Year, Month, and Day from Order_Date into new columns.

data["Order_Date_Day"] = data["Order_Date"].dt.day
data["Order_Date_Month"] = data["Order_Date"].dt.month
data["Order_Date_MonthName"] = data["Order_Date"].dt.month_name()
data["Order_Date_Year"] = data["Order_Date"].dt.year
```

```python
print(data.dtypes)
```

# 2. Data Selection and Filtering

```python
# Retrieve all orders placed in March 2024.

Orders_2024 = data[(data["Order_Date_Year"] == 2024) & (data["Order_Date_Month"] == 3)]
print(Orders_2024.count())
```

```python
# Find all orders where: Total_Amount > 1000 & Category is Electronics.

Electronics_Orders_Above1000 = data[(data["Total_Amount"] > 1000) & (data["Category"] == "Electronics")]
print(Electronics_Orders_Above1000.count())
```

```python
# Select only Completed orders.

Completed_Orders = data[(data["Order_Status"] == "Completed")]
print(Completed_Orders.count())
```

```python
# Retrieve orders placed from Mumbai or Delhi.

Order_Mumbai_Delhi = data[(data["City"] == "Mumbai") | (data["City"] == "Delhi")]
print(Order_Mumbai_Delhi.count())
```

# 3. Data Manipulation and Feature Engineering

```python
# Create a new column Discounted_Amount:
# Apply a 10 percent discount on orders where Total_Amount > 8000.

data["Discounted_Amount"] = data["Total_Amount"].apply(lambda x: x * 0.90 if x>8000 else x)
```

```python
# Create a column Price_Level based on Total_Amount:
# Low: Below 2000
# Medium: 2000 to 8000
# High: Above 8000

data["Price_Level"] = pd.cut(
    data["Total_Amount"],
    bins=[-float("inf"),2000,8000,float("inf")],
    labels=["Low","Medium","High"])
data[["Total_Amount","Discounted_Amount","Price_Level"]].head()
```

```python
# Create a new column Revenue_Per_Item:
# Revenue_Per_Item = Total_Amount / Quantity

data["Revenue_Per_Item"] = data["Total_Amount"] / data["Quantity"]
data[["Total_Amount","Quantity","Revenue_Per_Item"]].head()
```

# 4. Aggregation and Business Insights

```python
# Find the total revenue per Category.

revenue_per_category = data.groupby("Category")["Revenue_Per_Item"].sum().reset_index()
revenue_per_category.head()
```

```python
# Calculate the average order value per Payment_Method.

average_order = data.groupby("Payment_Method")["Total_Amount"].mean().reset_index()
average_order
```

```python
# Identify the top 5 cities by total sales.

Top5_City = data.groupby("City")["Revenue_Per_Item"].sum().reset_index().sort_values("Revenue_Per_Item", ascending=False).head(5)
Top5_City
```

```python
# Find the number of cancelled and returned orders per Category.

cancelled_returned_orders = data[data["Order_Status"].isin(["Cancelled", "Returned"])].groupby("Category").size().reset_index(name="Count")
cancelled_returned_orders
```

```python
# Determine the monthly sales trend.

monthly_sales_trend = data.groupby(["Order_Date_Month", "Order_Date_MonthName"])["Revenue_Per_Item"].sum().reset_index()
monthly_sales_trend
```
